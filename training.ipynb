```python
import pandas as pd
import numpy as np
from sklearn import linear_model
from sklearn import metrics
from sklearn.model_selection import train_test_split
from sklearn import svm
from sklearn.preprocessing import PolynomialFeatures
from sklearn.neural_network import MLPRegressor
```

```python
train_df=pd.read_csv('C:/Users/gurur/Desktop/Term1 Docs/ML/MLProject/trainData6stationsWithMon200.csv')
train_df.head()
```

```python
X=train_df[['Month','station elevation','Downward Long Wave Rad Flux','Downward Short Wave Rad Flux','Temperature','Upward Long wave Rad','Upward Long wave Rad Atmos','Upward Short Wave Rad','Max Temperature']].to_numpy()
Y=train_df['solar energy'].to_numpy()

train_x,test_x,train_y,test_y=train_test_split(X,Y,test_size=0.2,random_state=0)

lr=linear_model.Ridge(alpha=0.000000000001,normalize=True)
lr.fit(train_x,train_y)
ypred=lr.predict(test_x)
mse=metrics.mean_squared_error(test_y,ypred,squared=False)
print(mse)
```

```python
X=train_df[['Month','station elevation','Downward Long Wave Rad Flux','Downward Short Wave Rad Flux','Temperature','Upward Long wave Rad','Upward Long wave Rad Atmos','Upward Short Wave Rad','Max Temperature']].to_numpy()
Y=train_df['solar energy'].to_numpy()

poly=PolynomialFeatures(degree=3)
X=poly.fit_transform(X)

train_x,test_x,train_y,test_y=train_test_split(X,Y,test_size=0.2,random_state=0)

lr=linear_model.Ridge(alpha=0.000000000001,normalize=True)
lr.fit(train_x,train_y)
ypred=lr.predict(test_x)
mse=metrics.mean_squared_error(test_y,ypred,squared=False)
print(mse)
```
